```python
# Autoviz = it used to visualize the data in all possible ways mainly we can use this library for exploratory data analysis
from autoviz.AutoViz_Class import AutoViz_Class # here iam importing autoviz_class from autoviz library
AV=AutoViz_Class() # here iam initilizing the autoviz_class()
```

```python
filename="Titanic-Dataset.csv"
sep=","
dft=AV.AutoViz(
                filename,
                sep=",",
                depVar="",
                dfte=None,
                header=0,
                verbose=0,
                lowess=False,
                chart_format="svg",
                max_rows_analyzed=150000,
                max_cols_analyzed=30,
)
```
